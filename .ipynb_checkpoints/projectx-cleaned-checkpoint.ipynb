````python
#translates images to option play 
def get_response_image(text) :
    if text == 0:
        return "No image available"
    if not isinstance(text, str):
        print(text)
        text = text[0]
    try:
        response = client.chat.completions.create(model="gpt-4o", messages=[
                            {
                    "role": "system",
                    "content": 
"You are parsing tweets to interpret and synthesize information about stock plays. Remember that we want to note the direction as long/short based on if the play is being opened or closed. E.g. if the image refers to opening a position on a call option, the trade is long. If the image refers to the closing of a call option, the trade is also long, since we are closing a long position."

                            },
                            {
                              "role": "user",
                              "content": [
                                {"type": "text", "text": "What kind of stock purchase is this image describing? Please also specify if it is ultimately long or short, and whether the poster is opening or closing the position. Keep in mind if you are closing an originally long position, the direction should still be marked as long, as the original play was long. If a position is being closed, please track sentiment as if the stock play was being opened."},
                                {
                                  "type": "image_url",
                                  "image_url": {
                                    "url": text,
                                  },
                                },
                              ],
                            }
                          ])
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(e)
        pass
````

````python
def dynamic_prompting(text, sys_prompt, user_prompt):
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {
                    "role": "system",
                    "content": sys_prompt
                },
                {
                    "role": "user",
                    "content": [
                        {"type": "text", "text": user_prompt},
                        {"type": "text", "text": text}
                    ]
                }
            ],
        )
        
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(e)
        pass
````

````python
def initialize_webdriver():
    options = Options()
    options.add_argument("--headless")  # Run in background without opening a browser window
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-extensions")
    options.add_argument("--disable-popup-blocking")
    options.add_argument("--disable-dev-tools")
    options.add_argument('blink-settings=imagesEnabled=false')

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    return driver
````

````python
def get_jpg_url(driver, link):
    max_attempts = 3
    initial_wait = 0.5
    max_wait = 10

    for attempt in range(max_attempts):
        url = None
        try:
            if isinstance(link, list):  # Extract first element if URL is a list
                for u in link:
                    if "twitter" in u:
                        url = u
                        break
            elif not link:
                print("No Link Provided")
                return None
            else:
                url = link

            if not url:
                return None

            url = f'https://{url}'
            print(f"Attempting to access URL: {url}")

            driver.get(url)
            # Wait for potential redirects and the page to stabilize
            WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.TAG_NAME, 'article')))
            time.sleep(1)  # Extended sleep to ensure the page is loaded

            images = driver.find_elements(By.TAG_NAME, "img")
            for img in images:
                img_src = img.get_attribute('src')
                if 'media' in img_src:
                    print("img_src: " + img_src)
                    return img_src

            print("No media images found on the page")
            return None

        except Exception as e:
            print(f"General error processing URL {url}: {e}")
            time.sleep(60)

    print("Maximum attempts reached, aborting")
    return None
````

````python
#this block of code actually pulls tweets
driver = initialize_webdriver()
for i in range(len(heisenberg_tweets)):
    try:
        heisenberg_tweets.loc[i, 'jpg_url'] = get_jpg_url(driver, heisenberg_tweets.loc[i, 'image_url'])
        if i % 80 == 0 and i != 0:
            driver.quit()
            print("sleepy time")
            time.sleep(90)
            driver = initialize_webdriver()
    except Exception as e:
        print(f"Error processing row {i}: {e}")

driver.quit()
````

````python
#new
sys_prompt = """You are parsing tweets to interpret and synthesize information about stock plays. Reference examples as a guide to understand the format of the output.
Examples:
Text: $PARA Closed In 13.11 (yesterday)Out 13.24+1%+$65 profitJust trying to reduce long exposure heading into tomorrow. TRANSCRIBED IMAGE DATA: The image describes the sale of stock 500 shares PARA at $13.24
https://t.co/GpCKwDrfky 
Correct Output: { "Image Data": "This image describes a stock sale transaction, not an option play like a call or a put. Specifically, it details the sale of 500 shares of PARA (which is the ticker symbol for a stock) at an average fill price of $13.2401. It's a limit order set to sell at $13.24. The negative quantity (-500) indicates that shares are being sold rather than purchased.", "Position": "Close", "Ticker": "PARA", "Direction": "Long", "Type of Play": "Stock Purchase Sale", "Quantity": "-500 shares", "Expiry": "N/A", "Strike Price": "N/A", "Total Money In/Out": "+$6,620.05" }

Text: $GOOG Closed In .32 credit (Apr 30th)Out .05 debit+$135 profitSmall. Got dangerously close. But never worried. TRANSCRIBED IMAGE DATA: The image describes a vertical options spread for Google (GOOG) stock, specifically a put spread. Here are the details:- **Type of Spread**: Vertical Put Spread- **Underlying Asset**: GOOG (Google)- **Expiry Date**: May 3, 2024- **Strike Prices**: 165 (bought) and 162.5 (sold)Details of the trade:- **Quantity**: 5 contracts- **Total Cost**: Since you are paying 0.05 per contract for 5 contracts, the total debit is 5 * 0.05 = $25.In this trade, you are buying 5 put options at a strike price of 165 and selling 5 put options at a strike price of 162.5 for the same expiration date of May 3, 2024. This creates a vertical spread (specifically a bear put spread), aiming to profit from a decrease in the price of the underlying stock, Google, down to or below the lower strike price of 162.5.
Correct Output: { "Image Data": "The image describes a vertical options spread for Google (GOOG) stock, specifically a put spread. Here are the details:- Type of Spread: Vertical Put Spread- Underlying Asset: GOOG (Google)- Expiry Date: May 3, 2024- Strike Prices: 165 (bought) and 162.5 (sold)Details of the trade:- Quantity: 5 contracts- Total Cost: Since you are paying 0.05 per contract for 5 contracts, the total debit is 5 * 0.05 = $25.In this trade, you are buying 5 put options at a strike price of 165 and selling 5 put options at a strike price of 162.5 for the same expiration date of May 3, 2024. This creates a vertical spread (specifically a bear put spread), aiming to profit from a decrease in the price of the underlying stock, Google, down to or below the lower strike price of 162.5.", "Position": "Close", "Ticker": "GOOG", "Direction": "Short", "Type of Play": "Vertical Put Spread", "Quantity": "5 contracts", "Expiry": "May 3, 2024", "Strike Price": "165 (bought), 162.5 (sold)", "Total Money In/Out": "+$135". }

Text: $DIS Open Earnings playRisk $1,670 to make $330.Bmo, implied about a 7pt move. https://t.co/nyMZGS23OZ  TRANSCRIBED IMAGE DATA: The image describes a vertical put spread (also known as a bull put spread). This is a type of options strategy that involves selling one put option and buying another put option at a lower strike price but with the same expiration date.Here's the breakdown of the trade based on the image:- This is for the stock with the ticker symbol "DIS" (Walt Disney Company), with options expiring on May 10, 2024.- The vertical spread involves the 108 and 106 strike prices, meaning you are dealing with 108/106 put options.- The strategy specified is a put vertical spread:  - Selling 10 put options at the 108 strike price.  - Buying 10 put options at the 106 strike price.- The prices for the transactions are:  - Sold (shorted) the 108 strike put options at $0.86 each.  - Bought (long) the 106 strike put options at $0.53 each.- The net credit received for the spread is $0.33 per share (since options typically represent 100 shares, the total net credit is $33 per contract).In summary, the strategy is a bull put spread where you hope the price of Disney (DIS) stays above the higher strike price (108) by the expiration date so that both options expire worthless, and you keep the credit received upfront.
Correct Output: { "Image Data": "The image describes a vertical put spread (also known as a bull put spread). This is a type of options strategy that involves selling one put option and buying another put option at a lower strike price but with the same expiration date.Here's the breakdown of the trade based on the image:- This is for the stock with the ticker symbol 'DIS' (Walt Disney Company), with options expiring on May 10, 2024.- The vertical spread involves the 108 and 106 strike prices, meaning you are dealing with 108/106 put options.- The strategy specified is a put vertical spread: - Selling 10 put options at the 108 strike price. - Buying 10 put options at the 106 strike price.- The prices for the transactions are: - Sold (shorted) the 108 strike put options at $0.86 each. - Bought (long) the 106 strike put options at $0.53 each.- The net credit received for the spread is $0.33 per share (since options typically represent 100 shares, the total net credit is $33 per contract).In summary, the strategy is a bull put spread where you hope the price of Disney (DIS) stays above the higher strike price (108) by the expiration date so that both options expire worthless, and you keep the credit received upfront.", "Position": "Open", "Ticker": "DIS", "Direction": "Long", "Type of Play": "Vertical Put Spread", "Quantity": "10 contracts", "Expiry": "May 10, 2024", "Strike Price": "108 (sold), 106 (bought)", "Total Money In/Out": "+$330" }

Text: \\\\\\\\$CRM Closed\\\\n\\\\nIn 3.45 (yesterday)\\\\n\\\\nOut 5.85 \\\\n\\\\n+70%\\\\n+\\\\\\\\$240 https://t.co/knM6wjQLrJ  TRANSCRIBED IMAGE DATA: The image describes a options stock transaction for a call option contract on CRM (presumably Salesforce's ticker symbol).\\\\n\\\\nHere are the details:\\\\n- CRM 100 (Weeklys) 7 JUN 24 (7) 220 C\\\\n  - CRM: The stock ticker symbol for Salesforce.\\\\n  - 100: Number of shares per options contract.\\\\n  - Weeklys: Refers to the weekly expiring options.\\\\n  - 7 JUN 24: Expiration date of the option is June 7, 2024.\\\\n  - 220 C: This is a call option with a strike price of \\\\\\\\$220.\\\\n\\\\nTransaction details:\\\\n- -1 indicates the sale of one option contract.\\\\n- SELL @5.85 LMT: The sale was executed at a limit price of \\\\\\\\$5.85 per share.\\\\n- Trade Fills section confirms 1 contract was sold at the price of \\\\\\\\$5.85 on 5/31/24 at 6:40 AM.\\\\n\\\\nThis transaction represents a short call position (selling a call). The poster is opening the position by selling a call option. As such, it is ultimately a short position.

"""


user_prompt = "Can you retrieve the relevant information about the stock play being described in this tweet? Please return a dictionary that contains the following fields: Image Data, Position, Ticker, Direction, Short, Type of Play, Vertical Put Spread, Quantity, Expiry, Strike Price, Total Money In/Out. If the tweet references multiple tickers, only provide the first one. Do not include any markdown. Position is usually indicated in the beginning of the text (open or closed) " 

import requests
from PIL import Image
from io import BytesIO
import pandas as pd

def validate_and_process_image(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        
        # Check if image size is below 20 MB
        if len(response.content) > 20 * 1024 * 1024:
            raise ValueError("Image size exceeds 20 MB")

        # Check image format
        image = Image.open(BytesIO(response.content))
        if image.format.lower() not in ['png', 'jpeg', 'gif', 'webp']:
            raise ValueError("Unsupported image format")
        
        return response.content.decode('utf-8')  # Decode to string
    except Exception as e:
        print(f"Error processing image from {url}: {e}")
        return None

ht_image_only['image_response'] = ht_image_only['jpg_url'].apply(lambda x: validate_and_process_image(x) if x else None)
ht_image_only['image_response'] = ht_image_only['image_response'].apply(lambda x: x.replace('$', '\\$') if x else None)
ht_image_only['text'] = ht_image_only['text'].apply(lambda x: x.replace('$', '\\$') if pd.notna(x) else x)
ht_image_only['full_response'] = ht_image_only['text'].astype(str) + '  TRANSCRIBED IMAGE DATA: ' + ht_image_only['image_response'].astype(str)
ht_image_only['result'] = ht_image_only['full_response'].apply(lambda text: dynamic_prompting(text, sys_prompt, user_prompt))
````

````python
plays = pd.DataFrame(columns=['open_created_at', 'close_created_at', 'stock_ticker', 'Direction'])

# Dictionary to store open positions
open_positions = {}

# Loop through the dataframe
for index, row in ht_image_only.iterrows():
    ticker_direction = (row['Ticker'], row['Type of Play'])
    if row['Position'] == 'Open':
        open_positions[ticker_direction] = row['created_at']
    elif row['Position'] == 'Close':
        if ticker_direction in open_positions:
            plays = plays.append({
                'open_created_at': open_positions[ticker_direction],
                'close_created_at': row['created_at'],
                'stock_ticker': row['Ticker'],
                'direction': row['Direction']
            }, ignore_index=True)
            # Remove the matched open position
            del open_positions[ticker_direction]
        else:
            print(f"No matching open position found for: {ticker_direction}")
````
